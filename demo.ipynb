# TCSDG — quick-start demo

This notebook provides a **small-scale** running demonstration of the TCSDG algorithm using two bundled, publicly available example datasets:

* `example_data/classification_grape_cultivar.csv`; a three-class classification task based on grape cultivar identification. This dataset corresponds to the UCI Wine dataset, which contains chemical measurements of wines derived from three cultivars: https://archive.ics.uci.edu/dataset/109/wine.
* `example_data/regression_diabetes.csv`; a continuous-target regression task based on the scikit-learn Diabetes dataset: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html.

The purpose of this notebook is to verify that the pipeline runs end to end and to illustrate how to call the library functions. All core functionality is implemented in `sdg_functions.py`; this notebook only imports and invokes those functions.


> **Heads-up.** The first run downloads the TabICL teacher model and may take a few minutes. To keep the demo fast we use a reduced grid (2 seeds, ratios 1x/2x, one training fraction) and skip the benchmark SDG algorithms.

## 1. Setup

Install the dependencies once (from the repository root):

pip install -r requirements.txt

````python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
from sdg_functions import TcsdgConfig, run_task, summarize

print("Imports OK")
````

## 2. Classification demo

Load the example classification dataset and inspect it.

````python
df_clf = pd.read_csv("example_data/classification_grape_cultivar.csv")

print("shape:", df_clf.shape)
print("class counts:")
print(df_clf["cultivar"].value_counts())
df_clf.head()
````

Run a small TCSDG benchmark. `synthcity_plugins=[]` runs **TCSDG only**; add plugin names (e.g. `["ctgan"]`) to also benchmark SDG methods.

````python
cfg = TcsdgConfig()   # default hybrid hyper-parameters

res_clf = run_task(
    task="classification",
    df=df_clf,
    target_col="cultivar",
    cfg=cfg,
    seeds=[2, 3],            # full default: first 10 primes
    ratios=[1, 2],           # full default: [1, 2, 4, 8]
    train_fractions=[0.70],  # full default: [0.70, 0.30]
    synthcity_plugins=[],    # [] = TCSDG only; e.g. ["ctgan"] adds a baseline
)

summarize(res_clf, seeds=[2, 3])
````

## 3. Regression demo

The same workflow for a continuous target.

````python
df_reg = pd.read_csv("example_data/regression_diabetes.csv")

print("shape:", df_reg.shape)
df_reg.head()
````

````python
res_reg = run_task(
    task="regression",
    df=df_reg,
    target_col="target",
    cfg=cfg,
    seeds=[2, 3],
    ratios=[1, 2],
    train_fractions=[0.70],
    synthcity_plugins=[],
)

summarize(res_reg, seeds=[2, 3])
````
